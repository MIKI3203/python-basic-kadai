```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}は大人です。")
        else:
            print(f"{self.name}は大人ではありません。")


humans = [
    Human("侍太郎", 36),
    Human("侍花子", 18),
    Human("侍次郎", 20)
]

for human in humans:
    human.check_adult()
```
